```python
onehot = pd.get_dummies(completedataframe[['category']], prefix="", prefix_sep="")
pd.set_option("display.max_rows", None, "display.max_columns", None)
onehot['name'] = completedataframe['name']
onehot['lat'] = completedataframe['lat']
onehot['lng'] = completedataframe['lng']
onehot = onehot[['name','lat','lng','Restaurant','Store','Food Shop','Coffee Shop','Bar','Hotel','Gym','Pharmacy','Park','Theater']]
grouped = onehot.groupby(['name','lat','lng']).sum().reset_index()
grouped


```

```python
from sklearn.cluster import KMeans
from sklearn import preprocessing


kclusters = 5
clustering = grouped.drop(['name','lat','lng'], 1)
cols = clustering.columns

std = preprocessing.StandardScaler()
np_scaled = std.fit_transform(clustering)
clustering = pd.DataFrame(np_scaled, columns = cols)
clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(clustering)
kmeans.labels_[0:4] 
kmeans

finaldata = grouped
finaldata.insert(0, 'Cluster Labels', kmeans.labels_)
finaldata.to_csv('finaldata.csv')
finaldata

```

```python
import numpy as np
import matplotlib.cm as cm
import matplotlib.colors as colors

map_clusters = folium.Map(location=[8.983333, -79.516670], zoom_start=10)
finaldata['Cluster Labels'] = finaldata['Cluster Labels'].astype(int) 

x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

markers_colors = []
for lat, lon, poi, cluster in zip(finaldata['lat'], finaldata['lng'], finaldata['name'], finaldata['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
